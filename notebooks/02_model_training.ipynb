## Disease Prediction System

With hyperparameter tuning using GridSearchCV and Stratified K-Fold cross-validation.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from src.data_loader import load_dataset
from src.preprocessing import PreprocessingPipeline
from src.train import train_all_models, get_training_summary
from src.evaluate import (
    evaluate_all_models, create_comparison_table,
    generate_all_plots, get_best_model
)
from src.config import FIGURES_DIR, MODEL_RESULTS_DIR

print('All modules imported successfully!')
```

## 1. Heart Disease

```python
# Load and preprocess
heart_df = load_dataset('heart')
print(f"Heart Disease dataset: {heart_df.shape}")
print(f"Target distribution:\n{heart_df['target'].value_counts()}")

# Preprocessing
heart_pipeline = PreprocessingPipeline()
X_train_h, X_test_h, y_train_h, y_test_h = heart_pipeline.fit_transform(heart_df)

print(f"\nPreprocessing Report:")
for key, val in heart_pipeline.get_report().items():
    print(f"  {key}: {val}")
```

```python
# Train all models
heart_models = train_all_models(
    X_train_h, y_train_h, 'heart', tune_hyperparameters=True
)

# Training summary
heart_summary = get_training_summary(heart_models)
display(heart_summary)
```

```python
# Evaluate all models
heart_results = evaluate_all_models(heart_models, X_test_h, y_test_h, 'heart')

# Comparison table
heart_comparison = create_comparison_table(heart_results)
display(heart_comparison)

# Best model
best_heart = get_best_model(heart_results)
print(f"\n🏆 Best model for Heart Disease: {best_heart}")
```

```python
# Generate evaluation plots
generate_all_plots(heart_results, y_test_h, 'Heart Disease')
print('All heart disease plots generated!')
```
